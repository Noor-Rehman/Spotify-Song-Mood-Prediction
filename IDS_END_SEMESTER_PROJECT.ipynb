```python
from graphviz import Digraph
from IPython.display import Image, display
import math

# Create a project workflow diagram
dot = Digraph(format='png', engine='neato', comment='Spotify Mood Prediction Workflow')

# Number of nodes
nodes = [
    'Start',
    'Data Collection',
    'Data Preprocessing',
    'Data Visualization',
    'Clustering Analysis',
    'Mood Prediction',
    'Insights',
    'Impact'
]

# Define styles for nodes
node_styles = {
    'Start': {'color': 'darkgreen', 'style': 'filled', 'shape': 'ellipse', 'fontcolor': 'white'},
    'Data Collection': {'color': 'skyblue', 'style': 'filled', 'shape': 'box'},
    'Data Preprocessing': {'color': 'dodgerblue', 'style': 'filled', 'shape': 'box'},
    'Data Visualization': {'color': 'gold', 'style': 'filled', 'shape': 'box'},
    'Clustering Analysis': {'color': 'orange', 'style': 'filled', 'shape': 'box'},
    'Mood Prediction': {'color': 'coral', 'style': 'filled', 'shape': 'box'},
    'Insights': {'color': 'lightgreen', 'style': 'filled', 'shape': 'ellipse'},
    'Impact': {'color': 'purple', 'style': 'filled', 'shape': 'ellipse', 'fontcolor': 'white'}
}

# Position the nodes in a circle
radius = 2.0  # Radius of the circle
for i, node in enumerate(nodes):
    angle = 2 * math.pi * i / len(nodes)
    x = radius * math.cos(angle)
    y = radius * math.sin(angle)
    dot.node(node, node, pos=f"{x},{y}!", **node_styles.get(node, {}))

# Add edges to represent the workflow
edges = [
    ('Start', 'Data Collection'),
    ('Data Collection', 'Data Preprocessing'),
    ('Data Preprocessing', 'Data Visualization'),
    ('Data Visualization', 'Clustering Analysis'),
    ('Clustering Analysis', 'Mood Prediction'),
    ('Mood Prediction', 'Insights'),
    ('Insights', 'Impact')
]

for edge in edges:
    dot.edge(*edge, color='gray', penwidth='2.0', arrowhead='vee')

# Render and display the diagram
dot.attr(overlap='false', splines='curved')  # Smooth edges
dot.render('spotify_project_workflow', cleanup=True)
display(Image(filename='spotify_project_workflow.png'))
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import time
import pandas as pd
from requests.exceptions import HTTPError
from spotipy.exceptions import SpotifyException

# Set up your credentials
client_id = 'a4075561f9424bdb97caca578c753399'
client_secret = '003caf549c1d4571bf3dce87dc970b5e'

# Authentication
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))
```

```python
artists_data = []
query = 'a'  # A common keyword to search for a wide range of artists
limit = 50  # Maximum results per request
offset = 0  # Start from the first result

while len(artists_data) < 1000:
    results = sp.search(q=query, type='artist', limit=limit, offset=offset)
    artists = results['artists']['items']

    if not artists:
        break  # Stop if no more artists are found

    for artist in artists:
        artist_id = artist['id']
        artist_name = artist['name']
        genres = artist['genres']
        popularity = artist['popularity']
        followers = artist['followers']['total']
        artist_url = artist['external_urls']['spotify']

        # Fetch the artist's top tracks
        top_tracks = sp.artist_top_tracks(artist_id)['tracks']

        for track in top_tracks[:10]:  # Limit to 10 tracks per artist
            track_id = track['id']
            track_name = track['name']
            album_name = track['album']['name']
            release_date = track['album']['release_date']
            duration_ms = track['duration_ms']
            explicit = track['explicit']
            popularity = track['popularity']

            # Fetch audio features for the track
            features = sp.audio_features(track_id)[0]

            if features:  # Ensure features are returned
                danceability = features['danceability']
                energy = features['energy']
                key = features['key']
                loudness = features['loudness']
                mode = features['mode']
                speechiness = features['speechiness']
                acousticness = features['acousticness']
                instrumentalness = features['instrumentalness']
                liveness = features['liveness']
                valence = features['valence']
                tempo = features['tempo']

                data = {
                    'artist_name': artist_name,
                    'genres': genres,
                    'followers': followers,
                    'artist_popularity': popularity,
                    'artist_url': artist_url,
                    'track_name': track_name,
                    'album_name': album_name,
                    'release_date': release_date,
                    'duration_ms': duration_ms,
                    'explicit': explicit,
                    'track_popularity': popularity,
                    'danceability': danceability,
                    'energy': energy,
                    'key': key,
                    'loudness': loudness,
                    'mode': mode,
                    'speechiness': speechiness,
                    'acousticness': acousticness,
                    'instrumentalness': instrumentalness,
                    'liveness': liveness,
                    'valence': valence,
                    'tempo': tempo
                }

                artists_data.append(data)

    offset += limit  # Move to the next set of results
    time.sleep(0.1)  # Sleep briefly to avoid hitting rate limits

# Ensuring we only have 1,000 artists
artists_data = artists_data[:1000]

# Convert to DataFrame
df = pd.DataFrame(artists_data)
print(df)
```

```python

from sklearn.cluster import KMeans  #imports KMeans clustering algorithm (Partition data into Clusters on basis of Similarity )
wcss=[]                             #(With in cluster sum of squares) measures Squared average distance of each point in a cluster to the cluster's centroid
'''
The goal is to minimize WCSS meaning that points in each cluster are as close as possible to their centroids.
'''
#Loop to Compute WCSS for different Cluster Counts

for i in range(1,11):    #This loop iterates over a range of cluster numbers from 1 to 10, to find the elbow point which indicates the optimal number of clusters
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(features_scaled)
    wcss.append(kmeans.inertia_)

data=pd.DataFrame({'Number of Clusters':range(1,11),'WCSS':wcss})
fig=px.line(data,x='Number of Clusters',y='WCSS',color_discrete_sequence=['Green'],template='plotly_dark',title='<b>Elbow Curve')
fig.update_layout(title_x=0.5,xaxis_title=dict(text='Number of Clusters',font_size=17),yaxis_title=dict(text='WCSS',font_size=17))
fig.show()
```

```python
#fit the k-means with the chosen number of clusters
optimal_clusters=4               #Chose on basis of elbow point
kmeans=KMeans(n_clusters=optimal_clusters,init='k-means++',random_state=42)
clusters=kmeans.fit_predict(features_scaled)

df['clusters']=clusters          #Add the cluster labels to the rignal DataFrame.
```

```python
# function to assign moods based on all feature thresholds
def assign_mood(row):
    # Conditions for 'happy'
    if row['valence'] > 0.6 and row['energy'] > 0.5 and row['danceability'] > 0.5 and row['loudness'] > -5 and row['tempo'] > 100:
        return 'happy'
    # Conditions for 'sad'
    elif row['valence'] < 0.4 and row['energy'] < 0.4 and row['danceability'] < 0.4 and row['tempo'] < 90 and row['loudness'] < -10:
        return 'sad'
    # Conditions for 'energetic'
    elif row['energy'] > 0.7 and row['danceability'] > 0.7 and row['tempo'] > 120 and row['loudness'] > -5:
        return 'energetic'
    # Conditions for 'calm'
    elif row['acousticness'] > 0.7 and row['energy'] < 0.4 and row['tempo'] < 80 and row['loudness'] < -20 and row['danceability'] < 0.4:
        return 'calm'
    # Reclassify 'neutral' to 'happy' based on valence and energy
    elif row['valence'] > 0.5 and row['energy'] > 0.5:
        return 'happy'
    # Reclassify 'neutral' to 'energetic' based on energy and tempo
    elif row['energy'] > 0.5 and row['tempo'] > 100:
        return 'energetic'
    # Reclassify 'neutral' to 'sad' if valence and energy are both low
    elif row['valence'] <= 0.5 and row['energy'] <= 0.5:
        return 'sad'
    else:
        return 'calm'  # Default fallback is 'calm'
```

```python
# Filter features and target
features = df[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
               'instrumentalness', 'liveness', 'valence', 'tempo']]
df['mood'] = df.apply(assign_mood, axis=1)  # Apply the mood function
target = df['mood']
```

```python
# Encode target labels
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
target_encoded = label_encoder.fit_transform(target)
```

```python
# @title
from IPython.display import display, HTML

html_code = """

<div style="color:white;
            display: block;
            border-radius:10px;
            background-color:darkgreen;
            font-size:90%;
            font-family:Verdana;
            letter-spacing:0.5px;
            text-align: center;
            margin-top: 10px;">
    <h4 style='padding: 15px;
              color:white;
              margin: 0;'>
        MODEL TRAINING
    </h4>
</div>
"""

display(HTML(html_code))
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(features, target_encoded, test_size=0.2, random_state=42, stratify=target_encoded)
```

```python
# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Define classifier
model = SVC(kernel='rbf', C=1, gamma='scale', random_state=42)
```

```python
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
```

```python
# Cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=cv, scoring='accuracy')

print(f"Cross-Validation Accuracy Scores: {cv_scores}")
print(f"Mean CV Accuracy: {np.mean(cv_scores):.2f}")
```
